```python
import yfinance as yf
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Veri çekme
stock = "THYAO.IS"
start_date = datetime(2012, 2, 2)
end_date = datetime(2023, 12, 20)
bist30_data = yf.download(stock, start=start_date, end=end_date)
df = bist30_data['Close']

# Veriyi eğitim ve test setlerine bölme
train_size = int(len(df) * 0.80)
train, test = df[0:train_size], df[train_size:]

# SARIMA modelini eğitme
order = (5, 1, 1)  # (p, d, q)
seasonal_order = (1, 1, 1, 5)  # (P, D, Q, s)
model = SARIMAX(train, order=order, seasonal_order=(1, 1, 1, 5))
model_fit = model.fit()

# Gelecekteki fiyatları tahmin etme
future_steps = len(test)
predictions = model_fit.get_forecast(steps=future_steps).predicted_mean

# Görselleştirme
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Eğitim Verisi', color='blue')
plt.plot(test.index, test, label='Gerçek Veri', color='green')
plt.plot(test.index, predictions, label='Tahminler', color='red')
plt.title('Hisse Senedi Fiyat Tahmini - SARIMA Modeli')
plt.xlabel('Tarih')
plt.ylabel('Kapanış Fiyatları')
plt.legend()
plt.show()

# Hata hesaplama
mse = mean_squared_error(test, predictions)
print(f'Mean Squared Error (MSE): {mse}')
```

```python
import yfinance as yf
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Veri çekme
stock = "THYAO.IS"
start_date = datetime(2012, 2, 2)
end_date = datetime(2023, 12, 20)
bist30_data = yf.download(stock, start=start_date, end=end_date)
df = bist30_data['Close']

# Veriyi eğitim ve test setlerine bölme
train_size = int(len(df) * 0.80)
train, test = df[0:train_size], df[train_size:]

# SARIMA modelini eğitme
order = (5, 1, 1)  # (p, d, q)
seasonal_order = (1, 1, 1, 5)  # (P, D, Q, s)
model = SARIMAX(train, order=order, seasonal_order=(seasonal_order), enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit()

# Gelecekteki fiyatları tahmin etme
future_steps = len(test)
predictions = model_fit.get_forecast(steps=future_steps).predicted_mean

# Görselleştirme
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Eğitim Verisi', color='blue')
plt.plot(test.index, test, label='Gerçek Veri', color='green')
plt.plot(test.index, predictions, label='Tahminler', color='red')
plt.title('Hisse Senedi Fiyat Tahmini - SARIMA Modeli')
plt.xlabel('Tarih')
plt.ylabel('Kapanış Fiyatları')
plt.legend()
plt.show()

# Hata hesaplama
mse = mean_squared_error(test, predictions)
print(f'Mean Squared Error (MSE): {mse}')
```

```python
import yfinance as yf
from fbprophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Veri çekme
stock = "THYAO.IS"
start_date = "2012-02-02"
end_date = "2023-12-20"
bist30_data = yf.download(stock, start=start_date, end=end_date)
df = bist30_data['Close'].reset_index()

# Veriyi prophet uygun formata dönüştürme
df = df.rename(columns={'Date': 'ds', 'Close': 'y'})

# Prophet modelini oluşturma
model = Prophet()
model.fit(df)

# Gelecekteki fiyatları tahmin etme
future = model.make_future_dataframe(periods=365)  # 1 yıl sonrası için tahmin
forecast = model.predict(future)

# Tahmin sonuçlarını görselleştirme
fig = model.plot(forecast)
plt.title('Hisse Senedi Fiyat Tahmini - Prophet Modeli')
plt.xlabel('Tarih')
plt.ylabel('Kapanış Fiyatları')
plt.show()
```
