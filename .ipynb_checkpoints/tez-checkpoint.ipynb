```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense

# Veri çekme
stock = "THYAO.IS"
start_date = "2012-02-02"
end_date = "2023-12-02"
bist30_data = yf.download(stock, start=start_date, end=end_date)
df = bist30_data['Close'].reset_index()
prices = bist30_data["Close"].values.reshape(-1, 1)

# Veriyi normalize et
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# Veriyi train ve test setlere böl
train_size = int(len(prices_scaled) * 0.80)
train, test = prices_scaled[0:train_size, :], prices_scaled[train_size:len(prices_scaled), :]

# Veriyi modelin giriş formatına dönüştür
def create_dataset(dataset, time_steps=1):
    x, y = [], []
    for i in range(len(dataset) - time_steps - 1):
        a = dataset[i:(i + time_steps), 0]
        x.append(a)
        y.append(dataset[i + time_steps, 0])
    return np.array(x), np.array(y)

time_steps = 20
X_train, y_train = create_dataset(train, time_steps)
X_test, y_test = create_dataset(test, time_steps)

# Girişi yeniden şekillendir (samples, time_steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# CNN modelini oluştur
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(time_steps, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))

# Modeli derle
model.compile(optimizer='adam', loss='mean_squared_error')

# Modeli eğit
model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=2)

# Tahminleri yap
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Normalleştirmeyi geri al
train_predict = scaler.inverse_transform(train_predict)
y_train = scaler.inverse_transform([y_train])
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform([y_test])

# Eğitim setindeki hata
train_score = np.sqrt(mean_squared_error(y_train[0], train_predict[:, 0]))
print('Train Score: %.2f RMSE' % (train_score))

# Test setindeki hata
test_score = np.sqrt(mean_squared_error(y_test[0], test_predict[:, 0]))
print('Test Score: %.2f RMSE' % (test_score))

# Tahminleri görselleştir
train_predict_plot = np.empty_like(prices_scaled)
train_predict_plot[:, :] = np.nan
train_predict_plot[time_steps:len(train_predict) + time_steps, :] = train_predict

test_predict_plot = np.empty_like(prices_scaled)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict) + (time_steps * 2) + 1:len(prices_scaled) - 1, :] = test_predict

# Gerçek fiyatları görselleştir
plt.plot(scaler.inverse_transform(prices_scaled), label='True Prices')

# Eğitim seti tahminlerini görselleştir
plt.plot(train_predict_plot, label='Train Predictions')

# Test seti tahminlerini görselleştir
plt.plot(test_predict_plot, label='Test Predictions')

plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Veri çekme (yfinance kütüphanesinin yüklü olması gerekiyor)
import yfinance as yf

stock = "THYAO.IS"
start_date = "2012-02-02"
end_date = "2023-12-02"
bist30_data = yf.download(stock, start=start_date, end=end_date)
df = bist30_data['Close'].reset_index()
prices = bist30_data["Close"].values.reshape(-1, 1)

# Veriyi normalize et
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# Veriyi train ve test setlere böl
train_size = int(len(prices_scaled) * 0.80)
train, test = prices_scaled[0:train_size, :], prices_scaled[train_size:len(prices_scaled), :]

# Veriyi modelin giriş formatına dönüştür
def create_dataset(dataset, time_steps=1):
    x, y = [], []
    for i in range(len(dataset) - time_steps - 1):
        a = dataset[i:(i + time_steps), 0]
        x.append(a)
        y.append(dataset[i + time_steps, 0])
    return np.array(x), np.array(y)

time_steps = 20
X_train, y_train = create_dataset(train, time_steps)
X_test, y_test = create_dataset(test, time_steps)

# Girişi yeniden şekillendir (samples, time_steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# RNN modelini oluştur
model = Sequential()
model.add(SimpleRNN(units=50, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

# Modeli derle
model.compile(optimizer='adam', loss='mean_squared_error')

# Modeli eğit
model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=2)

# Tahminleri yap
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Normalleştirmeyi geri al
train_predict = scaler.inverse_transform(train_predict)
y_train = scaler.inverse_transform([y_train])
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform([y_test])

# Eğitim setindeki hata
train_score = np.sqrt(mean_squared_error(y_train[0], train_predict[:, 0]))
print('Train Score: %.2f RMSE' % (train_score))

# Test setindeki hata
test_score = np.sqrt(mean_squared_error(y_test[0], test_predict[:, 0]))
print('Test Score: %.2f RMSE' % (test_score))

# Tahminleri görselleştir
train_predict_plot = np.empty_like(prices_scaled)
train_predict_plot[:, :] = np.nan
train_predict_plot[time_steps:len(train_predict) + time_steps, :] = train_predict

test_predict_plot = np.empty_like(prices_scaled)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict) + (time_steps * 2) + 1:len(prices_scaled) - 1, :] = test_predict

# Gerçek fiyatları görselleştir
plt.plot(scaler.inverse_transform(prices_scaled), label='True Prices')

# Eğitim seti tahminlerini görselleştir
plt.plot(train_predict_plot, label='Train Predictions')

# Test seti tahminlerini görselleştir
plt.plot(test_predict_plot, label='Test Predictions')

plt.legend()
plt.show()
```

```python

# Gelecek 1 hafta için tahminlerde bulun
future_time_steps = 5  # Örneğin, 5 gün sonrasını tahmin edelim
future_predictions = []

last_week_data = prices_scaled[-time_steps:]  # Son 1 haftanın verilerini al
for _ in range(future_time_steps):
    input_data = last_week_data[-time_steps:].reshape(1, time_steps, 1)
    predicted_price = model.predict(input_data)
    future_predictions.append(predicted_price[0, 0])
    last_week_data = np.append(last_week_data, predicted_price, axis=0)

# Normalleştirmeyi geri al
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Gelecek 1 haftayı görselleştir
plt.figure(figsize=(10, 6))

# Gerçek fiyatları görselleştir
plt.plot(scaler.inverse_transform(prices_scaled), label='True Prices')

# Gelecek 1 hafta tahminlerini görselleştir
future_range = np.arange(len(prices_scaled), len(prices_scaled) + future_time_steps)
plt.plot(future_range, future_predictions, label='Future Predictions', linestyle='dashed', color='red')

plt.legend()
plt.show()
```
